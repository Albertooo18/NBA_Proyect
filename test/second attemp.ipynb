```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que ya tienes tus datos en un DataFrame llamado df
# Asegúrate de tener las columnas 'PTS' y 'MP' en tu DataFrame

# Seleccionar las estadísticas de interés
stats_seleccionadas = df[['PTS', 'MP']]

# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(stats_seleccionadas['PTS'], stats_seleccionadas['MP'], alpha=0.5)
plt.xlabel('Puntos por Juego (PTS)')
plt.ylabel('Minutos por Juego (MP)')
plt.title('Relación entre Puntos y Minutos por Juego de los Jugadores')
plt.grid(True)  # Agregar cuadrícula al gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Supongamos que ya tienes tus datos en un DataFrame llamado df
# Asegúrate de tener las columnas correspondientes en tu DataFrame

# Gráfico 1: Histograma de Puntos por Juego (PTS)
plt.figure(figsize=(10, 6))
plt.hist(df['PTS'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Puntos por Juego (PTS)')
plt.ylabel('Frecuencia')
plt.title('Histograma de Puntos por Juego')
plt.grid(True)
plt.show()
```

```python
# Gráfico 2: Gráfico de Barras de Asistencias por Juego (AST)
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['AST'], color='green', alpha=0.7)
plt.xlabel('Jugador')
plt.ylabel('Asistencias por Juego (AST)')
plt.title('Asistencias por Juego de los Jugadores')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# Gráfico 3: Gráfico de Cajas de Bloqueos por Juego (BLK)
plt.figure(figsize=(10, 6))
plt.boxplot(df['BLK'], vert=False)
plt.xlabel('Bloqueos por Juego (BLK)')
plt.title('Distribución de Bloqueos por Juego')
plt.grid(True)
plt.show()
```

```python
# Gráfico 4: Diagrama de Violín de Porcentaje de Tiros de Campo Exitosos (FG%)
plt.figure(figsize=(10, 6))
plt.violinplot(df['FG%'], vert=False, showmeans=True)
plt.xlabel('Porcentaje de Tiros de Campo Exitosos (FG%)')
plt.title('Distribución de FG%')
plt.grid(True)
plt.show()
```

```python
# Gráfico 5: Gráfico de Barras Horizontales de Edad (Age)
plt.figure(figsize=(10, 6))
plt.barh(df.index, df['Age'], color='orange', alpha=0.7)
plt.ylabel('Jugador')
plt.xlabel('Edad')
plt.title('Edad de los Jugadores')
plt.grid(True)
plt.show()
```
