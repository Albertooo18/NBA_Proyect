```python
import pandas as pd

df = pd.read_csv('../databases/nba_salaries_&_stats3.csv')
```

Relación entre Puntos Promedio y Salario por Posición en la NBA

```python
import matplotlib.pyplot as plt

# Calcular el salario promedio por posición
average_salary_by_position = df.groupby('Position')['Salary'].mean()

# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))

for position in df['Position'].unique():
    plt.scatter(df[df['Position'] == position]['PTS'], df[df['Position'] == position]['Salary'] / 1000000, label=position)

plt.xlabel('Puntos Promedio')
plt.ylabel('Salario Promedio (en millones de dólares)')
plt.title('Relación entre Puntos Promedio y Salario por Posición')
plt.legend(title='Posición')
plt.grid(True)
plt.tight_layout()
plt.savefig('Relación entre Puntos Promedio y Salario por Posición en la NBA')
plt.show()
```

Gráfico de Barras de Estadísticas ofensivas por Posición

```python
import matplotlib.pyplot as plt

# Crear gráfico de barras de estadísticas ofensivas por posición
offensive_stats = df.groupby('Position')[['PTS', 'FG%', '3P%', 'FT%']].mean()

plt.figure(figsize=(10, 6))
offensive_stats.plot(kind='bar', stacked=False)
plt.xlabel('Posición')
plt.ylabel('Promedio')
plt.title('Promedio de Estadísticas Ofensivas por Posición')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Estadísticas')
plt.tight_layout()
plt.savefig('Gráfico de Barras de Estadísticas ofensivas por Posición')
plt.show()
```

Gráfico de Barras de Estadísticas Defensivas por Posición

```python
import matplotlib.pyplot as plt

# Crear gráfico de barras de estadísticas defensivas por posición
defensive_stats = df.groupby('Position')[['STL', 'BLK', 'TOV']].mean()

plt.figure(figsize=(10, 6))
defensive_stats.plot(kind='bar', stacked=False)
plt.xlabel('Posición')
plt.ylabel('Promedio')
plt.title('Promedio de Estadísticas Defensivas por Posición')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Estadísticas')
plt.tight_layout()
plt.savefig('Gráfico de Barras de Estadísticas Defensivas por Posición')
plt.show()
```

Salario Promedio por Posición en la NBA

```python
import matplotlib.pyplot as plt

# Calcular el salario promedio de todos los jugadores
average_salary_all = df['Salary'].mean()

# Configurar los datos para el gráfico
positions = df['Position'].unique()
positions.sort()
salaries = df.groupby('Position')['Salary'].mean()
salaries = salaries.apply(lambda x: x / 1000000)  # Convertir los salarios a millones
salaries_mean = average_salary_all / 1000000  # Convertir el salario promedio a millones

# Configurar los ejes y los datos de la gráfica
fig, ax = plt.subplots(figsize=(12, 6))
positions_numeric = range(len(positions))

# Dibujar las barras de salario por posición
ax.bar(positions_numeric, salaries, color='skyblue')

# Dibujar la línea vertical para el salario promedio
ax.axhline(y=salaries_mean, color='red', linestyle='--', label='Salario Promedio')

# Configurar los ejes y la leyenda
ax.set_xticks(positions_numeric)
ax.set_xticklabels(positions, rotation=45)
ax.set_ylabel('Salario Promedio (en millones de dólares)')
ax.set_xlabel('Posición')
ax.set_title('Salario Promedio por Posición')
ax.legend()

# Most
plt.savefig('Salario Promedio por Posición en la NBA')
```
