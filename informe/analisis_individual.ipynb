Importación de Pandas y Lectura de la Base de Datos

```python
import pandas as pd

df = pd.read_csv('../databases/nba_salaries_&_stats3.csv')
```

Gráfico de Distribución de Puntos Anotados por Juego

```python
import matplotlib.pyplot as plt

# Crear histograma de distribución de salario
plt.figure(figsize=(10, 6))
plt.hist(df['PTS'], bins=20, color='orange', alpha=0.7, label='Puntos por Juego')
plt.hist(df['Salary'] / 1000000, bins=20, color='skyblue', alpha=0.7, label='Salario (en millones)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Distribución de Puntos por Juego y Salarios')
plt.grid(True)
plt.legend()
plt.savefig('Gráfico de Distribución de Puntos Anotados por Juego.png')
plt.show()
```

Gráfico de dispersión de altura vs salario

```python
import matplotlib.pyplot as plt

# Dividir la columna de salario para ajustarla a una escala manejable
df['Salario_dividido'] = df['Salary'] / 1000000  # Dividir por un millón para obtener el salario en millones de dólares

# Gráfico de dispersión de salario vs altura
plt.figure(figsize=(10, 6))
plt.scatter(df['Altura'], df['Salario_dividido'], color='skyblue', alpha=0.7)
plt.xlabel('Altura (pies)')
plt.ylabel('Salario (en millones de dólares)')
plt.title('Relación entre Altura y Salario')
plt.grid(True)
plt.savefig('Gráfico de dispersión de altura vs salario.png')
plt.show()

```

Gráfico de dispersión de peso vs salario

```python
# Gráfico de dispersión de salario vs peso
plt.figure(figsize=(10, 6))
plt.scatter(df['Peso'], df['Salario_dividido'], color='salmon', alpha=0.7)
plt.xlabel('Peso (libras)')
plt.ylabel('Salario (en millones de dólares)')
plt.title('Relación entre Peso y Salario')
plt.grid(True)
plt.savefig('Gráfico de dispersión de peso vs salario.png')
plt.show()
```
